```python
# Лабараторная работа 1

# Ермакова Арирна Сергеевна, Циб-251

# Вариант 9: Стек. Проверка строки на палиндром (игнорируя пробелы).

# Класс `Stack[T]` (src/my_stack.py)
#  Методы:
 # push(item) - O(1)
 # pop() - O(1)
 # peek() - O(1)
 # is_empty() - O(1)
 # size() - O(1)

 # Сложность алгоритма:
  # Время: O(n), где n — длина очищенной строки
  # Память: O(n) для хранения символов в стеке

# src/my_stack.py
from typing import Any, Generic, TypeVar
T = TypeVar('T')
class Stack(Generic[T]):
    def __init__(self) -> None:
        self._items: list[T] = []
    def push(self, item: T) -> None:
        self._items.append(item)
    def pop(self) -> T:
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self._items.pop()
    def peek(self) -> T:
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self._items[-1]
    def is_empty(self) -> bool:
        return len(self._items) == 0
    def size(self) -> int:
        return len(self._items)
    def __len__(self) -> int:
        return self.size()
```

```python
# main.py (логически)

def is_palindrome(text: str) -> bool:
    cleaned = ''.join(ch.lower() for ch in text if ch != ' ')
    if len(cleaned) <= 1:
        return True

    stack = Stack[str]()
    for ch in cleaned:
        stack.push(ch)

    for ch in cleaned:
        if ch != stack.pop():
            return False
    return True
```

```python
def check_palindrome_interactive():
    print("Проверка строки на палиндром")
    print("Введите текст для проверки или 'выход' для завершения:\n")

    while True:
        user_input = input("Введите строку: ").strip()

        if user_input.lower() in ('выход', 'exit', 'quit'):
            print("Работа завершена.")
            break

        if not user_input:
            print("Строка пустая. Введите текст или 'выход'.\n")
            continue

        result = is_palindrome(user_input)
        if result:
            print(f"'{user_input}' — это палиндром!\n")
        else:
            print(f"'{user_input}' — это НЕ палиндром.\n")

check_palindrome_interactive()
```
